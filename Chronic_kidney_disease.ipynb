```python
# Importing Standard Libraries
import os

#Importing EDA Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')


df = pd.read_csv('/content/ckd/kidney_disease_dataset.csv')
df_cop = df.copy()
```

```python
df['Target'] = df['Target'].replace({"No_Disease": "No", "Low_Risk": "Yes", "High_Risk": "Yes", 'Severe_Disease': 'Yes', 'Moderate_Risk': "Yes"})
```

```python
df['Red blood cells in urine'] = df['Red blood cells in urine'].map({'normal':0, 'abnormal':1})
```

```python
df['Pus cells in urine'] = df['Pus cells in urine'].map({'normal':0, 'abnormal':1})
df['Bacteria in urine'] = df['Bacteria in urine'].map({'not present': 0, 'present':1})
```

```python
YNcols = ["Hypertension (yes/no)",
"Diabetes mellitus (yes/no)",
"Coronary artery disease (yes/no)",
"Pedal edema (yes/no)",
"Anemia (yes/no)",
"Family history of chronic kidney disease",
'Smoking status']

for feature in YNcols:
  df[feature] = df[feature].map({'yes': 1, 'no': 0})
```

```python
for cols in df.columns:
  if(df[cols].dtypes == 'object'):
    print(cols)
```

```python
df['Physical activity level'] = df['Physical activity level'].map({'low':0, 'moderate':1, 'high':2})
```

```python
df['Urinary sediment microscopy results']= df['Urinary sediment microscopy results'].map({'normal':0, 'abnormal':1})
```

```python
df['Pus cell clumps in urine'] = df['Pus cell clumps in urine'].map({'not present': 0, 'present':1})
```

```python
df['Appetite (good/poor)'] = df['Appetite (good/poor)'].map({'good':0, 'poor':1})
df['Target'] = df['Target'].map({'No':0, 'Yes':1})
```

```python
# Machine Learning

# Importing Machine Learning Libraries

# since, our model is needed to be accurate and fast we are using XGboost for the model training of that
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

x = df.iloc[:,:-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
X_train.shape, X_test.shape
```

```python
model = XGBClassifier()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# classification report
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred) * 100)
```

```python
# hyper parameter tuning
# Importing Grid Search Cv for the best params
from sklearn.model_selection import GridSearchCV

param_grid = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': [50, 100, 200],
}

search = GridSearchCV(estimator=model, param_grid=param_grid, verbose=3, n_jobs=2, cv=5)
search.fit(X_train, y_train)
```

```python
search.best_params_
```

```python
best_model = search.best_estimator_
```

```python
best_model
```

```python
y_pred_best = best_model.predict(X_test)

print(classification_report(y_test, y_pred_best))
print(confusion_matrix(y_test, y_pred_best))
print(accuracy_score(y_test, y_pred_best) * 100)
```

Insights >> The Model had 75% accuracy before the Hyper parameter tuning but after that it has 80% of accuracy
